```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("../Dataset/CAvideos.csv")
```

```python
data
```

```python
data_50 = data.head(50)
```

```python
#relplot will give a scatterplot by default
sns.set()
sns.relplot(x="views", y="likes", data=data, hue='category_id', size='comment_count')

#Provides built-in features
#x and y values are automatically given labels as opposed to matplotlib
```

```python
sns.set()
sns.relplot(x="views", y="likes", data=data_50, hue='channel_title')
```

```python
#Linear Regression Plot
sns.regplot(x="views", y="likes", data=data);
```

```python
#Categorical Plots
sns.boxplot(x='category_id',y='views', data=data_50)

#Looking at Q
```

```python
#Can use with Matplotlib
import matplotlib.pyplot as plt
f, axes = plt.subplots(1, 2, sharey=True, figsize=(6, 4))
sns.boxplot(x='category_id',y='views', data=data_50, ax=axes[0])
sns.scatterplot(x='likes', y='views', hue='category_id', data=data_50, ax=axes[1])
```
